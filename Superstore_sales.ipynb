```python
pivot_table = df.pivot_table(values="Sales", index="Category", columns="Region", aggfunc="sum")
plt.figure(figsize=(8,5))
sns.heatmap(pivot_table,annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales by Category and Region")
plt.show()
```

```python
df["year"]=df["Order_Date"].dt.year
df["year"]
```

```python
sales_by_year=df.groupby("year")["Sales"].sum()
sales_by_year
```

```python
sales_by_year.plot(kind='pie',autopct="%.0f%%")
plt.title("Sales by Year")
plt.ylabel(" ")
plt.legend()
plt.show()
```

```python
year_sales = df.groupby("year")["Sales"].sum()
plt.pie(year_sales, labels=year_sales.index, 
        autopct=lambda p: '{:.0f}'.format(p * sum(year_sales)/100)) 
plt.title("Sales by year (Values)")
plt.show()
```
